```python
#load the dataset
import pandas as pd
vet = pd.read_csv(r"C:\Users\Esther\Desktop\git_ML\vertebrate(4).csv")
```

```python
vet
```

```python
vet['Class']=vet['Class'].replace(['reptiles','fishes','amphibians','birds'],'non-mammals')
```

```python
vet
```

```python
pd.crosstab([vet['Warm-blooded'],vet['Gives Birth']], vet['Class']) # Use the 'Class' column as values and the DataFrame `vet` for the index
```

```python
#decision trees
from sklearn import tree
```

```python
y=vet['Class']
```

```python
y
```

```python
x=vet.drop(['Name','Class'],axis=1)
```

```python
x
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy',max_depth=3)
```

```python
clf
```

```python
clf=clf.fit(x,y)
```

```python
clf
```

```python
import pydotplus
```

```python
from IPython.display import Image
from graphviz import Source
```

```python
dot_vet=tree.export_graphviz(clf,feature_names=x.columns,class_names=['mammals','non-mammals'], filled=True,out_file=None)
```

```python
dot_vet
```

```python
#convert to graph
graph=pydotplus.graph_from_dot_data(dot_vet)
```
